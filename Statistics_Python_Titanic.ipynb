## Descriptive Statistics

````python
df['age'].isnull().sum()
````

````python
# Summary statistics
df.describe()
````

````python
# Mean, Median, Mode for Age
print('Mean age:', df['age'].mean())
print('Median age:', df['age'].median())
print('Mode age:', df['age'].mode().values)
````

````python
# Minimum Age, Maximum Age, Age Range
print('Minimum age:', df['age'].min())
print('Maximum age:', df['age'].max())
print('Age range:', df['age'].max() - df['age'].min())
````

````python
# Standard Deviation, Variance, Range
print('Standard Deviation:', df['age'].std())
print('Variance:', df['age'].var())
print('Range:', df['age'].max() - df['age'].min())
````

## Visualizations

````python
# Histogram and KDE for Age
sns.histplot(df['age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
````

````python
# Boxplot for Age
sns.boxplot(x=df['age'])
plt.title('Age Boxplot')
plt.show()
````

## Mini Project: Class vs Age and Fare

````python
# Average Age and Fare by Pclass
df.groupby('pclass')[['age', 'fare']].mean()
````

````python
# Boxplots for Age and Fare by Class
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x='pclass', y='age', data=df, ax=axes[0])
axes[0].set_title('Age by Passenger Class')

sns.boxplot(x='pclass', y='fare', data=df, ax=axes[1])
axes[1].set_title('Fare by Passenger Class')
plt.tight_layout()
plt.show()
````
